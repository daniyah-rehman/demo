```python
import pandas as pd
df= pd.read_csv('Banking_Marketing.csv')#Add your own file path to run this cell
df= pd.DataFrame(df)
df
```

```python
# Dimensions of the dataset
num_rows, num_cols = 41199, 21
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_cols}")
```

```python
# Variable Identification
from sklearn.compose import make_column_selector as selector

numerical_columns_selector = selector(dtype_exclude=object)
categorical_columns_selector = selector(dtype_include=object)
numeric_vars =numerical_columns_selector(df)
categorical_vars =categorical_columns_selector(df)

print("Numeric Variables:")
print(numeric_vars)

print("\nCategorical Variables:")
print(categorical_vars)
```

```python
# Data Types
data_types =df.dtypes
print("\nData Types:")
print(data_types)
```

```python
# Summary Statistics for Numeric Variables
summary_stats =df.describe()
print("\nSummary Statistics for Numeric Variables:")
summary_stats
```

```python
# Check for missing values
missing_values =df.isnull().sum()

# Display the missing values
print("Missing Values:\n", missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extract numerical and categorical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include='object').columns

## Continuous Variables

# Histograms and Box Plots
for col in numerical_cols:
    plt.figure(figsize=(10, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    plt.hist(df[col], bins=30, color='orange', edgecolor='black')
    plt.title(f'Histogram of {col}')

    # Box Plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color='red')
    plt.title(f'Box Plot of {col}')

    plt.tight_layout()
    plt.show()
```
